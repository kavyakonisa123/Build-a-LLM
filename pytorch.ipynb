## Scalars, vectors , matrices, and tensors

1. Scalar (0D Tensor): A single number. No direction, no dimensions

* Examples include : Temperature = 25°C, Loss value = 0.87, Learning rate = 0.001

* In Deep Learning we Use for: Loss values, Hyperparameters, Single outputs

2. Vector (1D Tensor): A list of numbers.It has magnitude + direction
* Examples: [2,5,7], Word embedding, Feature vector
* In Deep Learning we represents: Words (embeddings), Features of an input, Hidden states
3. Matrix (2D Tensor): A 2D grid of numbers (rows × columns)
* Example: [[1,2,3],[4,5,6]]
* In Deep Learning it used for: Batch of vectors, Weight matrices, Input data
* 3D Tensor → Image --> (Height × Width × Channels)
4. Tensor : A generalization of scalars, vectors, and matrices
Can have any number of dimensions
* 4D Tensor → Batch of images --> (Batch × Height × Width × Channels)

### What is Reshape

Reshape : Changes tensor shape without changing data

`Total elements must stay the same`

````python
print(tensor2D.reshape(2,3))
````

### What is View?
* Similar to reshape 

BUT:
* Does NOT copy data
* Works only if tensor is contiguous in memory

````python
print(tensor2D.view(2,3))
````

## Transpose - Non contigous data

Transpose swaps rows and columns.
* Before: (6 × 3)
* After: (3 × 6)


This allows dot products between every pair of input vectors.

````python
print(tensor2D.shape)
print(tensor2D.T.shape)
````

````python
print(tensor2D.view(2,3))
print(tensor2D.T.view(2,3)) # This will work because the transposed tensor has the correct shape (2,3)
````

````python
print(tensor2D.T.view(3,2)) # This will raise an error because the transposed tensor has a different shape (3,2) vs (2,3)  
````

````python
print(tensor2D.T.reshape(3,2)) 
````

````python
x = torch.arange(6).reshape(2,3)
print(x)
x.T.view(6)   
````

````python
x = torch.arange(6).reshape(3,2)
print(x)
x.T.view(2,3)  # works
````

````python
print(x.T.reshape(2,3))  # also works because reshape can handle the change in shape as long as the total number of elements remains the same
````

```
* Shape → how data is arranged
* Reshape → rearrange safely
* View → rearrange efficiently (memory-level)
```

## Multiplications

What This Means

* Each element: result[i][j]=xi⋅xj
	​
* Measures similarity between vectors
* Used heavily in self-attention (QKᵀ)

````python
print(tensor2d.matmul(tensor2d.T))  # This will perform matrix multiplication between tensor2d and its transpose
````

````python
tensor1 = torch.tensor([[1,2], [3,4], [5,6] , [8,9]])
tensor2 = torch.tensor([[10,11], [12,13], [14,15], [16,17]])
print(tensor1)
print(tensor1.shape)
print(tensor2.T)
print(tensor2.T.shape)
tensor3 = tensor1.matmul(tensor2.T)  # This will perform matrix multiplication between tensor
print(tensor3, tensor3.shape)
````

````python
tensor3 = tensor1 @(tensor2.T)  # This will perform matrix multiplication between tensor
print(tensor3, tensor3.shape)
````

## Seeing models as computational Graphs

### Logistic regression forward pass

````python
import torch.nn.functional as F
y = torch.tensor([1.0])
x1 = torch.tensor([1.1])
w1 = torch.tensor([2.2])
b = torch.tensor([0.0])
print(y, x1, w1, b)
````

````python
z= x1*w1 + b
print(z)
a = torch.sigmoid(z)
print(a)
loss = F.binary_cross_entropy(a, y)
print(loss)
````
