## Attention mechanisms

```python
import torch

# Your Journey starts with one step
inputs = torch.tensor(
    [[0.43, 0.15, 0.89], #x^1
    [0.55, 0.87, 0.66], #x^2
    [0.57, 0.85, 0.64], #x^3
    [0.22, 0.58, 0.33], #x^4
    [0.77, 0.25, 0.10], #x^5
    [0.05, 0.80, 0.55]] #x^6
)
```

```python
print(inputs.shape)
print(inputs[0].shape)
print(inputs.shape[0])
print(inputs.shape[1])
```

```python
query = inputs[1]
attention_scores_2 = torch.empty(inputs.shape[0])
for i, x_i in enumerate(inputs):
    attention_scores_2[i] = torch.dot(x_i, query)
print(attention_scores_2)
```

```python
print(attention_scores_2.sum())
attention_weights_2_tmp = attention_scores_2 / attention_scores_2.sum()
print("Attention Weights (Tmp):", attention_weights_2_tmp)
print("Sum of Attention Weights (Tmp):", attention_weights_2_tmp.sum())
```

```python
def softmax_v1(x):
    exp_x = torch.exp(x)
    return exp_x / exp_x.sum(dim=0)
```

```python
attention_weights_2_naive = softmax_v1(attention_scores_2)
print("Attention Weights (Naive):", attention_weights_2_naive)
print("Sum of Attention Weights (Naive):", attention_weights_2_naive.sum())
```

```python
attention_weights = torch.softmax(attention_scores_2, dim=0)
print("Attention Weights (PyTorch):", attention_weights)
print("Sum of Attention Weights (PyTorch):", attention_weights.sum())
```

```python
query = inputs[1]
context_vector_2 = torch.empty(query.shape)
# context_vector = torch.zeros(query.shape)
for i, x_i in enumerate(inputs):
    context_vector_2 += attention_weights[i] * x_i
print("Context Vector:", context_vector_2)
```

### computing attention weights for all input tokens

```python
attention_scores = torch.empty(inputs.shape[0],inputs.shape[0])
for i, x_i in enumerate(inputs):
    for j, x_j in enumerate(inputs):
        attention_scores[i][j] = torch.dot(x_i, x_j)

print(attention_scores)
```

```python

attention_scores= inputs @ inputs.T
print(attention_scores)
```

```python
attention_weights = torch.softmax(attention_scores, dim = -1)
print(attention_weights)
```

```python
all_context_vectors = attention_weights @ inputs
print(all_context_vectors)
```
